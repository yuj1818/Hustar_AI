# 1. Problem 1

- Using the features extracted from the cell, diagnose breast cancer.
    - Part 1: Pandas basics

- Settings

```python
import pandas as pd
import numpy as np
import seaborn as sns

import matplotlib.pyplot as plt

%matplotlib inline
```

## 1.1. Part 1: Pandas basics

### 1.1.1. Data Load

```python
Total_data = pd.read_csv('./breast_cancer.csv')
Total_data
```

__(a) Problems_1: Data_NaN values check__

```python
'''
각 columns 에 존재하는 NaN 수를 집계하라.  
'''

Total_data.isna().sum()
```

__(b) Problems_2: Data_Remove columns__

- id : No need for learning
<br>
- Unnamed: 32 : NaN column 

```python
'''
필요없는 columns(id, Unnamed: 32)를 삭제하라.
'''

cancer_data = Total_data.drop(columns=['id','Unnamed: 32'])
```

```python
cancer_data
```

### 1.1.2. DataLabeling

- Target : diagnosis column
<br>
- M = Cancer / B = Normal

```python
print("Total number of labels: {}".format(cancer_data.shape[0]))
print("Number of cancer: {}".format(cancer_data[cancer_data.diagnosis == 'M'].shape[0]))
print("Number of normal: {}".format(cancer_data[cancer_data.diagnosis == 'B'].shape[0]))
```

```python
cancer_label = pd.Categorical(cancer_data['diagnosis']).codes
```

```python
cancer_label = cancer_label.reshape(-1,1)
```

- cancer : 1, normal : 0

```python
cancer_label
```

```python
pd_num_label = pd.DataFrame(cancer_label, columns=['numerical_label'])
```

### 1.1.3. Data Combined

- Combine : cancer_data + pd_num_label

```python
pd_cancer_feature = pd.concat([cancer_data, pd_num_label], axis=1)
```

- Final data form

```python
pd_cancer_feature
```

- Shape Check

```python
pd_cancer_feature.shape
```

__(c) Problems_3: Data_Exploration & Visualization__

```python
pd_cancer_feature[['radius_mean', 'texture_mean', 'perimeter_mean',
       'area_mean', 'smoothness_mean', 'compactness_mean', 'concavity_mean',
       'concave points_mean', 'symmetry_mean', 'fractal_dimension_mean',
       'radius_se', 'texture_se', 'perimeter_se', 'area_se', 'smoothness_se',
       'compactness_se', 'concavity_se', 'concave points_se', 'symmetry_se',
       'fractal_dimension_se', 'radius_worst', 'texture_worst',
       'perimeter_worst', 'area_worst', 'smoothness_worst',
       'compactness_worst', 'concavity_worst', 'concave points_worst',
       'symmetry_worst', 'fractal_dimension_worst']].std()
```

- Perform the analysis on data
    - Using the describe()

```python
'''
아래의 표와 같이 주어진 데이터에 대해서
각 칼럼마다 count(데이터 수), mean 값, std 값, min 값 등을 구하여라.

- pandas의 describe() 기능 사용
'''
pd_cancer_feature.describe()
```

- check the distribution of the area_worst(maximum std)

```python
'''
데이터 분석으로 얻어진 결과에서 표준편차가 최대값을 가지는 area_worst 특징에 대해서 
아래와 같이 데이터 분포를 확인해라.
sns.distplot() 함수를 사용하라.
'''
sns.distplot(pd_cancer_feature['area_worst'])
```

__(d) Problems_4: Data_preprocessing__

- Remove outliers
    - Outliers : area worst > 3000

```python
pd_cancer_feature[pd_cancer_feature['area_worst']>3000]
```

```python
'''
area_worst의 데이터 분포를 확인하였을 때, area_worst가 3000 이상인 값은 이상치로 판단하였다.
area_worst가 3000 이상인 이상치 데이터를 제거해라.
'''
RemoveOutlier_cancer = pd_cancer_feature[pd_cancer_feature["area_worst"]<=3000]
RemoveOutlier_cancer
```

- Check

```python
plt.figure(figsize=(16, 8))
plt.title("[Distribution][Remove_outliers] area_worst", fontsize = 15)
sns.distplot(RemoveOutlier_cancer['area_worst'] , color="darkviolet")
plt.legend()
```

- Define Input/Target

```python
input_data = RemoveOutlier_cancer.drop(['diagnosis', 'numerical_label'], axis = 1)
output_data = RemoveOutlier_cancer['numerical_label']
```

- Normalization
    - min : 0 / max : 1

```python
from sklearn.preprocessing import minmax_scale
```

```python
input_MinMax_scaled = minmax_scale(input_data, axis=0, copy=True)
```

```python
input_data.columns
```

```python
input_data.index
```

- Results

```python
pd_input_scaled_data = pd.DataFrame(input_MinMax_scaled, columns=input_data.columns, index=input_data.index)
```

```python
pd_input_scaled_data
```

### 1.1.4. Data_Split_Train/Test

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(pd_input_scaled_data, output_data, test_size=0.3, random_state=1)
```

```python
print("Number of train data: {}".format(y_train.shape[0]))
print("Number of test data: {}".format(y_test.shape[0]))
```
