```python
import numpy as np
import cvxpy as cvx

f = np.array([[2],[3]])

A = np.array([[2, 2],
              [1, 2]])

b = np.array([[20],
              [14]])

lb = np.array([[1],
               [2]])

x = cvx.Variable([2,1])

obj = cvx.Minimize(-f.T @ x)
constraints = [A@x <= b, lb <= x]

prob = cvx.Problem(obj, constraints)
result = prob.solve()

print(x.value)
print(result)
```

# Exercise 2
- Find a value that satisfies the following formula using cvxpy method
<img src=attachment:image.png align="center" width="200" height="200">
- Hint(exercise 1 과 동일한 풀이)
    - 식을 행렬로 표현
    - x를 cvx.Variable((2,1))로 표현하고, objective와 constraints를 설정
    - cvx.Problem(objective, contraints)로 cvs문제 정의
    - solve()를 이용하여 계산

```python
f = np.array([[1],[2]])
H = np.array([[1/2, 0],[0, 0]])

A = np.array([[-2, -3],
              [3, 4],
              [2, 4]])

b = np.array([[-12],
              [80],
              [80]])

lb = np.array([[0],
               [0]])

x = cvx.Variable([2,1])

obj = cvx.Minimize(cvx.quad_form(x, H) + f.T@x)
constraints = [A@x <= b, lb <= x]

prob = cvx.Problem(obj, constraints)
result = prob.solve()

print(x.value)
print(result)
```

# Exercise 3
- Find a value that satisfies the following formula using gradient descent method


- Hint
    - 식을 quadratic function으로 표현
    - 임의의 x를 설정 (np.zeros() or np.ones() 등 이용)
    - Step size (즉, 업데이트 횟수) 설정
    - For 문을 이용하여 gradient descent계산

```python
H = np.matrix([[5, -2],
              [0, 5]])
g = -np.matrix([[4],[-4]])

x = np.zeros((2,1))
alpha = 0.2

for i in range(25):
    df = H*x + g
    x = x - alpha*df
    
print(x)
```
